```python
import qiskit
import numpy as np

import ising

```

```python
Nq = 3

circ = qiskit.QuantumCircuit(Nq)

circ.h(0)
circ.cx(0,1)
circ.cx(0,2)


circ.draw()
```

```python
from qiskit.quantum_info import Statevector

state = Statevector.from_int(1,2**Nq)

state = state.evolve(circ)

state.draw('hinton')
```
